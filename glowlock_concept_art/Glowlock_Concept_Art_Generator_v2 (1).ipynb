```python
# üìå STEP 2: Login to Hugging Face
from huggingface_hub import login
login()  # You‚Äôll be prompted to enter your Hugging Face token
```

```python
# üìå STEP 3: Load the Stable Diffusion Model
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16,
    use_safetensors=True
).to("cuda")
```

```python
# üìå STEP 4: Enter Your Prompt
prompt = "storybook illustration of a glowing mossy forest trail at dusk, magical and foggy, fantasy aesthetic, soft lighting"
```

```python
# üìå STEP 5: Generate the Image
image = pipe(prompt).images[0]
from IPython.display import display
display(image)
```

```python
# üìå STEP 6: (Optional) Save the Image
image.save("glowlock_concept_art.png")
```
